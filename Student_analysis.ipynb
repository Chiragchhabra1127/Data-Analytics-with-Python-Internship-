**DATA LOAD**











```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data load karein
df = pd.read_csv("student-mat.csv", sep=';')

# Pehli 5 rows dekhne ke liye
print(df.head())
```

**Data Cleaning**

```python
# Missing values check karein
print(df.isnull().sum())

# Duplicates check karein
print(f"Duplicates found: {df.duplicated().sum()}")
```

**ANALYSIS**

```python
# 1. Average Final Grade (G3)
avg_g3 = df['G3'].mean()
print(f"Average Final Grade: {avg_g3:.2f}")

# 2. Kitne students ne 15 se zyada score kiya?
high_scorers = df[df['G3'] > 15].shape[0]
print(f"Students scoring above 15: {high_scorers}")

# 3. Gender wise performance
gender_performance = df.groupby('sex')['G3'].mean()
print("\nAverage score by Gender:")
print(gender_performance)
```

**VISUALISATION (A).Distribution of Final Grades**














```python
plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=10, kde=True, color='blue')
plt.title('Distribution of Final Grades (G3)')
plt.show()
```

**VISUALISATION (B).Study Time VS Final Grades**














```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='studytime', y='G3', data=df)
plt.title('Study Time vs Final Grades')
plt.show()
```
